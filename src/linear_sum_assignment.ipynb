```python
from scipy.optimize import linear_sum_assignment
import numpy as np
import pandas as pd
```

```python
costTable = pd.read_csv("data/address/processed/costTable/svr/"+"0_15,6,5"+".csv", sep=",",header=None)
```

```python
df= pd.read_csv("data/address/processed/addressList/addressList0.csv", sep=",",usecols=[1])
row=df.address.to_list()
```

```python
column=["none"]
for adr in df.address:
    column.append(adr)
print(type(row))
```

```python
row_ind, col_ind = linear_sum_assignment(costTable)
print(col_ind)
```

```python
class Address:

    
    def __init__(self, address, nextAddress):
        self.address = address
        self.nextAddress = nextAddress
        
    def getAddress(self):
        return self.address
    
    def getNextAddress(self):
        return self.nextAddress
```

```python
ansList=[]
i=0
for assign in col_ind:
    ansList.append(Address(row[i],column[assign]))
    i+=1
```

```python
f = open("data/result/multi/move/svr/0/15,6,5.txt", 'w')
for ans in ansList:
        f.write(ans.getAddress())
        f.write("\r\n")
        f.write(ans.getNextAddress())
        f.write("\r\n")
        f.write("\r\n")
f.close()
```
