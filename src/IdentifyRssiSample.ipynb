```python
#本コードはidentifyRssiの試作品になります
import pandas as pd
import numpy as np
#このあたりはコマンドライン引数からもらう予定
capture = pd.read_csv("data/capture/single/move/csv/move0.csv", sep=",")
baseAddress = '58:b4:ad:58:e0:0c'
testAddress = '52:ae:43:5a:b9:6f'
```

```python
#学習用データ作成
#今回は末尾3パケットで回帰する
x_train = []
y = []
y_train = []
for line in range(len(capture)):
    if capture.address[line] == baseAddress:
        x_train.append(capture.time[line])
        y.append(capture.rssi[line])
    
        
for num in range(len(y)):
    if len(y) - num < 4:
        y_train.append(y[num])
```

```python
#モデル生成
from sklearn.linear_model import LinearRegression
clf = LinearRegression()
clf.fit(pd.DataFrame(x_train).tail(3),y_train)
```

```python
#テストデータ作成
#今回は戦闘3パケットでやる
x_test = []
y_test = []
for line in range(len(capture)):
    if capture.address[line] == testAddress and len(x_test) < 3:
        x_test.append(capture.time[line])
        y_test.append(capture.rssi[line])

print(x_test)
print(y_test)
```

```python
#実際に回帰する
clf.predict(pd.DataFrame(x_test))
```
