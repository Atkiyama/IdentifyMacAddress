```python
import pandas as pd
import numpy as np
C = ", " #"何回も入力するのが面倒だった

capture = pd.read_csv("data/capture/single/move/csv/move0.csv", sep=",")
address = '58:b4:ad:58:e0:0c'
```

```python
#試作
x = capture.time
y = capture.rssi
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression

clf = LogisticRegression()
clf.fit(pd.DataFrame(x),y)
capture.shape
```

```python
#学習用データ作成
x_train = []
y_train = []
for line in range(len(capture)):
    if capture.address[line] == address:
        x_train.append(capture.time[line])
        y_train.append(capture.rssi[line])
```

```python
#モデル生成
clf.fit(pd.DataFrame(x_train),y_train)
```

```python
testAddress = '52:ae:43:5a:b9:6f'
x_test = []
y_test = []
for line in range(len(capture)):
    if capture.address[line] == testAddress:
        x_test.append(capture.time[line])
        y_test.append(capture.rssi[line])
        
results = clf.predict(pd.DataFrame(x_test))

for result in range(len(results)):
   print(abs(results[result]-y_test[result]))
```
