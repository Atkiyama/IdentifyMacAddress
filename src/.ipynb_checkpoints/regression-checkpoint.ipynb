```python
#諸々のインポート
import sys
import pandas as pd
import numpy as np
import csv
from sklearn.linear_model import LinearRegression
```

```python
def convertOriginal():
    import subprocess
    original = 'data/result/multi/move/convertOriginal.csv'
    subprocess.run(('java', 'convertMacAddress/ConvertOriginal',original,'20','data/capture/single/move/txt/'),check=True)
    return original
```

```python
#アドレスのクラス
class Address:
    def __init__(self,advA,fTime,lTime,rssi,numPkt):
        self.advA = advA
        self.fTime = fTime
        self.lTime = lTime
        self.rssi = [rssi]
        self.numPkt = numPkt
        self.nextAdr = []
```

```python
def makeAddressList(original):
    #read csv
    csv_file = open(original,"r",errors="", newline="")
    f = csv.reader(csv_file, delimiter=",", doublequote=True, lineterminator="\r\n", quotechar='"', skipinitialspace=True)

    #headerskip
    header = next(f)
    #print(header)

    #advA,fTime,lTime,rssi,num_packet
    data =[]
    #前に出たアドレスならTrue
    ad_known = False
    #全体のパケット数
    #各アドレスごとの最初と最後のパケットの時間を記録
    for row in f:
    #アドレス，最初の時間，最後の時間，rssi,パケット数
        for adr in data:
            if adr.advA == row[0]:
                adr.lTime = row[1]
                adr.rssi.append(int(row[2]))
                adr.numPkt += 1
                ad_known = True
        if not ad_known:
            data.append(Address(row[0],row[1],row[1],int(row[2]),1))
        ad_known = False
    
    return data
    
```
